```python
user_books = book.pivot_table(index='UserID',
                                 columns='BookTitle',
                                 values='BookRating').reset_index(drop=True) 
```

```python
user_books
```

```python
user_books.index = book.UserID.unique() 
user_books.index
```

```python
user_books.fillna(0, inplace=True)
user_books
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_books.values,metric='cosine')
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = book.UserID.unique()
user_sim_df.columns = book.UserID.unique() 
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

# IBCF

```python
un_user_rating = user_books["The Kitchen God's Wife"]  
un_user_rating
```

```python
similar_to_un = user_books.corrwith(un_user_rating) 
similar_to_un
```

```python
corr_Toy = pd.DataFrame(similar_to_un, columns=['Correlation'])
corr_Toy.dropna(inplace=True)
corr_Toy.head()
```

```python
corr_Toy[corr_Toy['Correlation'] > 0].sort_values(by='Correlation', ascending=False).head(5)  
```
